```python
import matplotlib.pyplot as plt

x = [1, 2, 6, 18]
y = [3, 10, 12, 20]

fig, ax = plt.subplots()

ax.plot(x, y, 'go', label='Points', markerfacecolor='green')

ax.plot(x, y, 'r--', linewidth=2, label='Line')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

ax.set_xlim(0, 20)
ax.set_ylim(0, 25)

ax.legend()

plt.show()

```

```python
import matplotlib.pyplot as plt


temperature = [12, 14, 16, 18, 20, 22, 24]

sales = [100, 200, 250, 400, 300, 450, 500]


# Plot the data
plt.plot(temperature, sales, marker='o', linestyle='-', color='purple', markersize=8)

# Add labels and a title
plt.xlabel('Temperature (°C)')
plt.ylabel('Sales')
plt.title('Sales vs. Temperature')

# Optional: Add grid lines
plt.grid(True)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt


x_values = []
y_values = []


with open('data.txt', 'r') as file:
    for line in file:
        x, y = map(float, line.split())
        x_values.append(x)
        y_values.append(y)


plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', markersize=8)


plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Title of the Line Plot')


plt.show()

```

```python
import matplotlib.pyplot as plt


x1 = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 17, 20]


x2 = [1, 2, 3, 4, 5]
y2 = [5, 8, 7, 10, 12]



plt.plot(x1, y1, marker='o', linestyle='-', color='b', markersize=8, label='Line 1')


plt.plot(x2, y2, marker='s', linestyle='--', color='g', markersize=8, label='Line 2')


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines on the Same Plot')


plt.legend()


plt.grid(True)


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(1, 11)


y1 = x  
y2 = x**2  
y3 = np.sqrt(x)  


plt.figure(figsize=(8, 6))


plt.plot(x, y1, label='y = x', color='blue')


plt.plot(x, y2, label='y = x^2', color='red')


plt.plot(x, y3, label='y = sqrt(x)', color='green')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Multiple Functions on the Same Graph')
plt.legend()


plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np


groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]


x = np.arange(len(groups))


bar_width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))


bar1 = ax.bar(x - bar_width/2, means_men, bar_width, label='Men', color='green')
bar2 = ax.bar(x + bar_width/2, means_women, bar_width, label='Women', color='red')


ax.set_xlabel('Groups')
ax.set_ylabel('Means')
ax.set_title('Scores by Group and Gender')
ax.set_xticks(x)
ax.set_xticklabels(groups)
ax.legend()


plt.show()
```

```python
import matplotlib.pyplot as plt


languages = ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
popularity = [22.2, 17.6, 8.8, 8, 77, 6.7]

plt.figure(figsize=(8, 8))
plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink', 'lightblue'])

plt.title('Popularity of Programming Languages')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from the CSV file
data = pd.read_csv('medals.csv')

# Sort the data by gold medals in descending order and select the top 5 countries
top_countries = data.nlargest(5, 'gold_medal')

# Extract the country names and gold medal counts
countries = top_countries['country']
gold_medals = top_countries['gold_medal']

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(gold_medals, labels=countries, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink'])

# Add a title
plt.title('Gold Medal Achievements in 2016 Summer Olympics (Top 5 Countries)')

# Show the plot
plt.show()
```
